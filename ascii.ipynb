```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
file_path = "/Users/apple/Downloads/caltech-101/101_ObjectCategories/strawberry/image_0002.jpg"
```

```python
image = cv2.imread(file_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# #resize image to 256x256
image = cv2.resize(image, (1000, 1000))
plt.imshow(image)
```

```python
#convert image into grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
gray.shape
```

```python
ascii_array = list("$@B%8&WM#*oahkbdpqwmZO0QLCJUYXzcvunxrjft/\|()1[]?-_+~<>i!lI;")
# make array of every second element of ascii_array
ascii_array = ascii_array[::5]
ascii_array = ascii_array[::-1]
print(ascii_array)
```

```python
copy = gray.copy()

copy_int64= copy.astype(np.int64)
copy_int64 *= len(ascii_array)
copy_int64 //= 999

r,c = copy_int64.shape
array = [['' for i in range(c)] for j in range(r)]
for i in range(r):
    for j in range(c):
        array[i][j] = ascii_array[copy_int64[i][j] - 1]

```
